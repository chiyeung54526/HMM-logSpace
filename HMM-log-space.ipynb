```python
import numpy as np

# read the data and convert data into ['H','T','H'...] format
raw_data = open('test2.txt').readlines()
seq_temp = raw_data[0].split()
seq=[]
for i in seq_temp[0]:
    seq.append(i)

# Initialize probability matrix
init_Matrix=np.array([[0.5,0.5]])

# transition matrix
transition_Matrix=np.array([[0.7,0.3],[0.4,0.6]])

# emission matrix
emission_Matrix=np.array([[0.5,0.5],[0.8,0.2]])

# step 1 init
# prob_Matrix: store the probabilities in each steps
prob_Matrix =[]
# put initial probability value into prob_Matrix
if seq[0] == 'H':
    prob_Matrix.append([np.log(init_Matrix[0][0]) + np.log(emission_Matrix[0][0]), np.log(init_Matrix[0][0]) + np.log(emission_Matrix[1][0])])
# binary choice situation: =='T' if not equal to 'H'
else:
    prob_Matrix.append([np.log(init_Matrix[0][1]) + np.log(emission_Matrix[0][1]), np.log(init_Matrix[0][1]) + np.log(emission_Matrix[1][1])])

# remove first element 
temp = seq[1:]
# store the opt path
viterbi_Res = []
for i in range(len(temp)):
    prev_F, prev_L = prob_Matrix[-1]
    if temp[i] =='H':
        F_score = max(prev_F + np.log(transition_Matrix[0][0]) + np.log(emission_Matrix[0][0]), prev_L + np.log(transition_Matrix[1][0]) + np.log(emission_Matrix[0][0]))
        L_score = max(prev_F + np.log(transition_Matrix[0][1]) + np.log(emission_Matrix[1][0]), prev_L + np.log(transition_Matrix[1][1]) + np.log(emission_Matrix[1][0]))
        prob_Matrix.append([F_score, L_score])
# also a binary choice situation: 'T' if not 'H'
    else:
        F_score = max(prev_F + np.log(transition_Matrix[0][0]) + np.log(emission_Matrix[0][1]), prev_L + np.log(transition_Matrix[1][0]) + np.log(emission_Matrix[0][1]))
        L_score = max(prev_F + np.log(transition_Matrix[0][1]) + np.log(emission_Matrix[1][1]), prev_L + np.log(transition_Matrix[1][1]) + np.log(emission_Matrix[1][1]))
        prob_Matrix.append([F_score, L_score])
        
# convert into log space (natural logarithm)
# prob_Matrix_log = np.log(prob_Matrix)
# the probability of the most likely state path
# print(prob_Matrix_log)
print(prob_Matrix)
```

```python
# optimal state path 
viterbi_Res = []
# for prob in prob_Matrix_log:
for prob in prob_Matrix:
#if prob[0] > prob[1] means F_score> L_score, then F wins, else L wins
    if prob[0] > prob[1]:
        viterbi_Res.append('F')
    else:
        viterbi_Res.append('L')
opt_path = ''.join(viterbi_Res)
print('optimal state path:',opt_path)
```
